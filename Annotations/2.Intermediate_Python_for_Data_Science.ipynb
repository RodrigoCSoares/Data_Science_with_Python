# 1. Matplotlib

Matplotlib is a package for data visualization which is very important in Data Analysis. The better you understand your data, the better will be the information extracted.

To plot data you need to give the information you wanna plot and you can also configure the plotting, there are a differents plot fuction for differents type of plots (i.e., linear, scatter, images, subplots, histograms, paths, etc) 

```python
# Import a csv file with a huge data of countries population
# We'll talk about pandas package and csv files later
import pandas as pd
population = pd.read_csv("Data/TotalPopulationBySex.csv")

import numpy as np

# Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Select Brazil's population
brazil = population[population['Location'] == 'Brazil']

# Select the year and the population of Brazil
brYear = np.array(brazil['Time'])
brPop = np.array(brazil['PopTotal'])

# Make a line plot: year on the x-axis, pop on the y-axis
plt.plot(brYear[:150], brPop[:150])

# Display the plot with plt.show()
plt.show()
```

### Histograms
Give a nice view of the distribution of your data.

There are a lot of arguments you can specify to build a histogram, but the most importants are the list of values you want to plot and the bins. Bins tells Python how many bins the data should be divided (10 by default).

```python
# Import a csv file with a huge data of countries population
# We'll talk about pandas package and csv files later
import pandas as pd
players = pd.read_csv("Data/soccerPlayers.csv")

# Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Save the age
age = players['age']

# Create histogram of age data
plt.hist(age)

# Show and clean up plot
plt.show()
plt.clf()

# Build histogram with 15 bins
plt.hist(age, bins = 15)

# Show and clean up again
plt.show()
```

### Customization

Sometimes just plot will not solve the problem, you will need to customize your plot to make the data very clear and understandable.

There are different plot types and for each type there are many customizations like colors, shapes, labels, axis, etc.

Examples of customization methods:
- xlabel(): names the x label
- ylabel(): names the y label
- title(): give a title to the plot
- yticks(): change the text plotted in the y label

```python
# Import a csv file with a huge data of countries population
# We'll talk about pandas package and csv files later
import pandas as pd
population = pd.read_csv("Data/TotalPopulationBySex.csv")

import numpy as np

# Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Select Brazil's population
brazil = population[population['Location'] == 'Brazil']

# Select the year and the population of Brazil
brYear = np.array(brazil['Time'])
brPop = np.array(brazil['PopTotal'])

# Make a line plot: year on the x-axis, pop on the y-axis
plt.plot(brYear[:150], brPop[:150])

# Strings
xlab = 'Year'
ylab = 'Total population'
title = 'Brazil\'s population per year'

# Add axis 
plt.xlabel(xlab)
plt.ylabel(ylab)

# Add title
plt.title(title)

# Display the plot with plt.show()
plt.show()
```

# 2. Dictionaries and Pandas 

Dictionaries are a good alternative to lists, the difference between list and dictionary is that dictionaries has "customized" index (dictionaries index = key value). Key values aren't numeric ordened, you can choose the key and give it a indexed value.

Dictionary sintax: dictionaryName = {key : value, key2 : value2, keyN : valueN}

Let's see how it really works and the difference between lists and dictionaries.

```python
# This code will give the same output using a list and using a dictionary

# Create a list
material = ['book', 'pen', 'backpack', 'eraser']
cost = [10, 2, 25, 5]

# Print the book's cost
costIndex = material.index('book')
print("Print with list: " + str(cost[costIndex]))

# Create a dictionary
costMaterial = {'book' : 10, 
                'pen' : 2,
                'backpack' : 25,
                'eraser' : 5
               }

# Print the book's cost
print("Print with dictionary: " + str(costMaterial['book']))
```

```python
# This code show a dictionary's key
# Definition of dictionary
europe = {'spain':'madrid', 'france':'paris', 'germany':'berlin', 'norway':'oslo' }

# Print out the keys in europe
print(europe.keys())
```

Dictionary's keys should be immutable (objects that you can't change the value after it has been created), it means that a key should be a number type or a string.

```python
# This is a example of a valid dictionary
valid = {0:'1', True:'2', False:'3', 'string':'4'}
```

```python
# This is a example of a invalid dictionary
invalid = {['list', 'type']:0}
```

```python
# This code shows how to add, remove and update values in a dictionary
# With this code you can understand better why keys should be unique.

# Create a dictionary
costMaterial = {'book' : 10, 
                'pen' : 2,
                'backpack' : 25,
                'eraser' : 5
               }

# Add a new material
costMaterial['paper'] = 3

# Update a value
costMaterial['book'] = 15

#Remove a value
del(costMaterial['pen'])

# Print the book's cost
print(costMaterial)
```

```python
# This code shows how to know if a key is in the dictionary

# Create a dictionary
costMaterial = {'book' : 10, 
                'pen' : 2,
                'backpack' : 25,
                'eraser' : 5
               }

# Print if the key is in the dictionary or not
print('book' in costMaterial)
```

# 3. Pandas

Pandas is a package for high level manipulation tool and built on NumPy package.

Take a look at the DataFram documentation, a useful Pandas class: https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.html

It's very useful to save and manipulate a large number of data. Furthermore, DataFrames has X and Y labels which makes easier to find everything you need.

There are many different ways to create a DataFrame, lets see how to create using a dictionary.

With Pandas you can import tons of data using a external file (like a .CSV file), instead doing it mannualy.

```python
# Import Pandas package
import pandas as pd

# Create a dictionary
people = {
    "name":["John", "Ana", "Peter", "Rose"],
    "age":[20, 23, 18, 30],
    "country":["USA", "BR", "RU", "CH"]
}

# Create a DataFrame using the dictionary
pd_people = pd.DataFrame(people)

# Print the DataFrame
print(pd_people)

# Brake line
print('\n')

# Change the index
pd_people.index = ['A', 'B', 'C', 'D']

# Print the DataFrame
print(pd_people)
```

```python
# This code shows a DataFrame created from a CSV file
import pandas as pd
players = pd.read_csv("Data/soccerPlayers.csv", index_col = 0)

# Print the type of players
print(type(players))

# Collumn accessing the DataFrame, double square brackets keep the output as a DataFrame
print(players[['name', 'age']])
```

Square brackets has limited functionality, a better way to select the data in your DataFrame is using the methods loc() and iloc().

loc() is label-based and iloc() is integer position-based

```python
# This code shows a DataFrame created from a CSV file
import pandas as pd
players = pd.read_csv("Data/soccerPlayers.csv", index_col = 0)

# Selecting a player using the index column value
print(players.loc[[20801],['name', 'full_name', 'club']])
```

```python
# This code shows a DataFrame created from a CSV file
import pandas as pd
players = pd.read_csv("Data/soccerPlayers.csv", index_col = 0)

# Selecting all rows in a specific column
print(players.loc[ : ,['name', 'club']])
```

```python
# This code shows a DataFrame created from a CSV file
import pandas as pd
players = pd.read_csv("Data/soccerPlayers.csv", index_col = 0)

# Selecting a player using the integer position
print(players.iloc[[0], [0, 1]])
```
